```python
files = "aspire.js  fricative.js  input2.js  white.js".split()
```

```python
import os
import json
import numpy as np
data = {}
for file in files:
    n, _ = os.path.splitext(file)
    print(n)
    with open(file) as fp:
        data[n] = np.array(json.load(fp), dtype=np.float64)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[10, 10])
for i, (k, v) in enumerate(data.items()):
    plt.subplot(2, 2, i + 1)
    print(k, np.mean(v), np.std(v), np.max(v), np.min(v))
    plt.hist(v, bins=100)
    plt.title(k)
```

```python
plt.figure(figsize=[10, 10])
for i, (k, v) in enumerate(data.items()):
    plt.subplot(2, 2, i + 1)
    f = np.linspace(0, 44100 / 2, 2048)
    y = np.log(np.abs(np.fft.fft(v, 4096)[:2048]))
    plt.plot(f, y)
    plt.title(k)
plt.show()
```

```python
plt.figure(figsize=[10, 10])
for i, (k, v) in enumerate(data.items()):
    plt.subplot(2, 2, i + 1)
    f = np.linspace(0, 44100 / 2, 256)
    y = np.log(np.abs(np.fft.fft(v, 512)[:256]))
    plt.xlim(0, 2000)
    plt.plot(f, y)
    plt.title(k)
plt.show()
```
